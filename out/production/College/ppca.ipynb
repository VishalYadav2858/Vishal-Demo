```python
# Demonstrating Tuples and Lists

# Creating a tuple
my_tuple = (1, 2, 3, 4, 5)
print("Tuple:", my_tuple)

# Creating a list
my_list = [1, 2, 3, 4, 5]
print("List:", my_list)

# Differences between tuples and lists
print("\nDifferences:")
# Immutability of tuples
try:
    my_tuple[0] = 10
except TypeError as e:
    print("Tuples are immutable:", e)

# Mutability of lists
my_list[0] = 10
print("Lists are mutable:", my_list)

# Applications of tuples and lists
# Lists can be used to store homogeneous items
homogeneous_list = [1, 2, 3, 4, 5]
print("\nHomogeneous list:", homogeneous_list)

# Lists can be used to store heterogeneous items
heterogeneous_list = [1, "apple", 3.14, True]
print("Heterogeneous list:", heterogeneous_list)

# Tuples are used for fixed collections of items
coordinates = (10, 20)
print("\nCoordinates (tuple):", coordinates)

# Lists are used for dynamic collections of items
fruits = ["apple", "banana", "cherry"]
print("Fruits (list):", fruits)

# Adding elements to a list
fruits.append("date")
print("List after appending:", fruits)
```

```python
def binary_to_decimal(binary_str):
    decimal = 0
    power = 0
    for digit in reversed(binary_str):
        if digit == "1":
            decimal += 2 ** power
        power += 1
    return decimal

def decimal_to_binary(decimal_num):
    if decimal_num == 0:
        return "0"
    binary_str = ""
    while decimal_num > 0:
        binary_str = str(decimal_num % 2) + binary_str
        decimal_num //= 2
    return binary_str

# Input binary from user
binar_1 = input("Enter first binary number: ")
binar_2 = input("Enter second binary number: ")

# Convert binary to decimal value
decimal_1 = binary_to_decimal(binar_1)
decimal_2 = binary_to_decimal(binar_2)

# Performing addition and multiplication
sum_decimal = decimal_1 + decimal_2
mul_decimal = decimal_1 * decimal_2

# Converting back to binary
sum_bin = decimal_to_binary(sum_decimal)
mul_bin = decimal_to_binary(mul_decimal)

print("Sum in binary:", sum_bin)
print("Multiplication in binary:", mul_bin)
```

```python
# Function to convert binary string to decimal
def binary_to_decimal(binary_str):
    decimal = 0
    power = 0
    for digit in reversed(binary_str):
        if digit == "1":
            decimal += 2**power
        power += 1
    return decimal


# Function to convert decimal to binary string
def decimal_to_binary(decimal_num):
    if decimal_num == 0:
        return "0"
    binary_str = ""
    while decimal_num > 0:
        binary_str = str(decimal_num % 2) + binary_str
        decimal_num //= 2
    return binary_str


# Input binary from user
binary_1 = input("Enter first binary number: ")
binary_2 = input("Enter second binary number: ")

# Convert binary to decimal manually using functions
dec_1 = binary_to_decimal(binary_1)
sum_2 = binary_to_decimal(binary_2)

# Performing addition and multiplication
sum_dec = dec_1 + sum_2
mul_dec = dec_1 * sum_2

# Converting result back to binary
sum_bin = decimal_to_binary(sum_dec)
mul_bin = decimal_to_binary(mul_dec)

# Displaying Results
print(f"Decimal of {binary_1} = {dec_1}")
print(f"Decimal of {binary_2} = {sum_2}")
print(f"Addition of binary = {sum_bin} (decimal: {sum_dec})")
print(f"Multiplication of binary = {mul_bin} (decimal: {mul_dec})")
```
